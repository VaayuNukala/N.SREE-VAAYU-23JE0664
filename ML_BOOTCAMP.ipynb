```python
class LogisticRegression:

    def __init__(self, iterations, learning_rate):
        self.iterations=iterations
        self.learning_rate=learning_rate
        self.W=None
        self.b=None
        self.num_classes=None
        self.probs=None
        self.loss_history=[]

    def initialize_parameters(self,num_features):
        self.W=np.zeros(self.num_classes,num_features)
        self.b=np.zeros((self.num_classes,))

    def softmax(self, x):
        z=np.dot(x,self.W.T) +self.b
        max_z=np.max(z, axis=1,keepdims=True)
        exp_z=np.exp(z-max_z)
        self.probs=exp_z/np.sum(exp_z,axis=1,keepdims=True)
        return self.probs

    def one_hot_encode(self, y):
        return np.eye(self.num_classes)[y]

    def gradient_calc(self, x, y_one_hot):
        y_diff=self.probs-y_one_hot
        del_W=-np.dot(y_diff.T,x)
        del_b=-np.sum(y_diff.T,axis=1)
        return del_W,del_b

    def loss_function_calc(self, y_one_hot):
        epsilon = 1e-10
        loss = -np.sum(y_one_hot * np.log(self.probs + epsilon)) / len(y_one_hot)
        return loss

    def train(self, x, y):
        num_features=x.shape[1]
        self.num_classes=np.unique(y).shape[0]
        self.initialize_parameters(num_features)
        one_hot_y=self.one_hot_encode(y)

        for iteration in range(self.iterations):
            self.probs = self.softmax(x)
            del_W,del_b = self.gradient_calc(x, one_hot_y)

            self.W-=self.learning_rate*del_W/len(x)
            self.b-=self.learning_rate*del_b/len(x)

            loss=self.loss_function_calc(one_hot_y)
            self.loss_history.append(loss)

        plt.plot(range(self.iterations), self.loss_history)
        plt.xlabel('Iterations')
        plt.ylabel('Loss')
        plt.title('Loss over iterations')
        plt.show()

        return self.W, self.b
```

```python
def euclidean_distance(x1,x2):
  distance=np.zeros((x2.shape[0],x1.shape[0]))
  for i in range(x2.shape[0]):
    distance[i]=np.sqrt(np.sum((x1-x2[i,:])**2,axis=1))

  return distance
def normalize(x):
    mean = np.mean(x, axis=1, keepdims=True)
    sigma = np.std(x, axis=1, keepdims=True)
    x = np.where(sigma == 0, 0, (x - mean) / sigma)
    return x

```

```python
class KNN:
  def __init__(self,k):
    self.k=k


  def mode(self, arr):
    max_count=0
    mode_value=None
    for value in arr:
      count=sum(1 for v in arr if np.array_equal(v, value))
      if count>max_count:
        max_count=count
        mode_value=value
    return mode_value




  def predict(self,X,x,y):

    distance=euclidean_distance(X,x)
    k_nearest_labels=np.zeros((x.shape[0],self.k))

    distance = np.argsort(distance.T, axis=1)[:,:self.k]

    k_nearest_labels = [y[m] for m in distance]
    return k_nearest_labels

  def test(self,X,x,y):

    arr=self.predict(X,x,y)
    y_=np.zeros((x.shape[0],))
    for i in range(x.shape[0]):
      arr_=arr[i]
      y_[i]=self.mode(arr_)
    return y_

  def accuracy_(self,X,x,y,y_t):
    y_=self.test(X,x,y)


    acc = np.sum(y_== y_t) /len(y_t)
    return acc
```
